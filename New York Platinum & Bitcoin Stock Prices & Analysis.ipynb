# New York Platinum & Bitcoin Stock Prices

Take the raw New York platinum stock price data from quandl and display the head

```python
import quandl
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style


api_key = open('quandlapikey.txt','r').read() # Reading text file containing quandl api key

query1 = "JOHNMATT/PLAT"
df_Plat_Raw = quandl.get(query1,authtoken=api_key,start_date="2014-5-1")
df_Plat_Raw.rename(columns={'New York 9:30':'New York Raw Platinum Stock'},inplace=True)
df_Plat_Raw = df_Plat_Raw.loc[:,'New York Raw Platinum Stock':'New York Raw Platinum Stock']
df_Plat_Raw.dropna(inplace=True)
df_Plat_Raw.head()
```

Plot the raw New York platinum stock price data

```python
df_Plat_Raw.plot() # Plots graph
plt.ylabel('Stock Price')
plt.title("New York Platinum Raw Stock Price")
```

Display the monthly averaged New York platinum stock price data from quandl and display the head

```python
df_Plat_Average = quandl.get(query1,authtoken=api_key,start_date="2014-5-1")
df_Plat_Average.rename(columns={'New York 9:30':'New York Average Platinum Stock Price'},inplace=True)
df_Plat_Average = df_Plat_Average.loc[:,'New York Average Platinum Stock Price':'New York Average Platinum Stock Price']
df_Plat_Average.dropna(inplace=True)

df_Plat_Average = df_Plat_Average.resample("M").mean()
df_Plat_Average.head()
```

Plot the monthly averaged New York platinum stock price data

```python
df_Plat_Average.plot()
plt.ylabel('Average Price')

plt.title("New York Platinum Average Monthly Stock Price")
```

Display the monthly averaged price percent change for New York platinum stock prices and display the head

```python
df_Plat_Perc = quandl.get(query1,authtoken=api_key,start_date="2014-5-1")
df_Plat_Perc.rename(columns={'New York 9:30':'New York Percent Change Platinum Stock Price'},inplace=True)
df_Plat_Perc = df_Plat_Perc.loc[:,'New York Percent Change Platinum Stock Price':'New York Percent Change Platinum Stock Price']
df_Plat_Perc.dropna(inplace=True)

df_Plat_Perc = df_Plat_Perc.pct_change()
df_Plat_Perc.dropna(inplace=True)
df_Plat_Perc = df_Plat_Perc.resample('M').mean()
df_Plat_Perc.head()
```

Plot the monthly averaged price percent change for New York platinum stock prices

```python
df_Plat_Perc.plot()
plt.ylabel('Percent Change')
plt.title("New York Platinum Average Stock Percent Change")
```

Take the raw Bitcoin stock price data from quandl and display the head

```python
query2 = "BCHAIN/MKPRU"
df_Bit_Raw = quandl.get(query2,authtoken=api_key,start_date="2014-5-1")
df_Bit_Raw.rename(columns={'Value':'Bitcoin Raw Stock Price'},inplace=True)
df_Bit_Raw = df_Bit_Raw.loc[:,'Bitcoin Raw Stock Price':'Bitcoin Raw Stock Price']
df_Bit_Raw.dropna(inplace=True)
df_Bit_Raw.head()
```

Plot the raw Bitcoin stock price data

```python
df_Bit_Raw.plot() # Plots graph
plt.ylabel('Stock Price')
plt.title("Bitcoin Raw Stock Price")
```

Display the monthly averaged Bitcoin stock price data from quandl and display the head

```python
df_Bit_Average = quandl.get(query2,authtoken=api_key,start_date="2014-5-1")
df_Bit_Average.rename(columns={'Value':'Bitcoin Average Stock Price'},inplace=True)
df_Bit_Average = df_Bit_Average.loc[:,'Bitcoin Average Stock Price':'Bitcoin Average Stock Price']
df_Bit_Average.dropna(inplace=True)

df_Bit_Average = df_Bit_Average.resample("M").mean()
df_Bit_Average.head()
```

Plot the monthly averaged Bitcoin stock price data

```python
df_Bit_Average.plot()
plt.ylabel('Average Price')

plt.title("Bitcoin Average Monthly Stock Price")
```

Display the monthly averaged price percent change for Bitcoin stock prices and display the head

```python
df_Bit_Perc = quandl.get(query2,authtoken=api_key,start_date="2014-5-1")
df_Bit_Perc.rename(columns={'Value':'Bitcoin Percent Stock Price Percent Change'},inplace=True)
df_Bit_Perc = df_Bit_Perc.loc[:,'Bitcoin Percent Stock Price Percent Change':'Bitcoin Percent Stock Price Percent Change']
df_Bit_Perc.dropna(inplace=True)

df_Bit_Perc = df_Bit_Perc.pct_change()
df_Bit_Perc.dropna(inplace=True)
df_Bit_Perc = df_Bit_Perc.resample('M').mean()
df_Bit_Perc.head()
```

Plot the monthly averaged price percent change for Bitcoin stock prices

```python
df_Bit_Perc.plot()
plt.ylabel('Percent Change')
plt.title("Bitcoin Stock Percent Change")
```

Plot the raw stock prices of each stock together.

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

df_Plat_Raw.plot(ax=ax1)
df_Bit_Raw.plot(ax =ax1)


plt.title("Platinum & Bitcoin Raw Stock Prices")
plt.legend()
plt.show()
```

Plot the monthly averaged stock price of each stock together.

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

df_Plat_Average.plot(ax=ax1)
df_Bit_Average.plot(ax =ax1)

plt.title("Platinum & Bitcoin Average Stock Price")             
plt.legend()
plt.show()
```

Plot the monthly average price percent change of each stock together.

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

df_Plat_Perc.plot(ax=ax1)
df_Bit_Perc.plot(ax =ax1)

plt.title("Platinum & Bitcoin Stock Percentage Change")             
plt.legend()
plt.show()
```

Merge the data into one dataset (or table).

```python
df_Plat_Bit = pd.merge(df_Plat_Raw,df_Bit_Raw,on='Date') # Merges the two datasets together
df_Plat_Bit.dropna(inplace=True)
print(df_Plat_Bit.head())
```

Create a correlation table to see the relationship of each stock when compared to each other.

```python
df_Plat_Bit_correlation = df_Plat_Bit.corr() # Creates a correlation table for the two stocks
print(df_Plat_Bit_correlation)
```

## Conclusion

The two stocks (Platinum and Bitcoin) have a negative relationship with each other according to the correlation table results. Therefore, these two stocks do not trend in the same direction.
